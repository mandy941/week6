## 1. What proportion of diamonds are between .30 and 1.08 carats?
#define the pandas and numpy
#read the url from github
#create a datafarm that calculates the proportion of diamonds in column 'carat' that satisfy the requirement

```python
#definde pandas and numpy
import pandas as pd
import numpy as np
# use read_csv function to load the csv from spcific websites.
df = pd.read_csv('https://raw.githubusercontent.com/mandy941/week6/master/diamonds.csv')
#calculate the value of column 'carat' that between 0.3 and 1.08 divides the total numble of rows.
porp = pd.value_counts((df['carat'] <= 1.08) & (df['carat'] >= 0.30)) / df.shape[0]
print(porp)
```

According to the result, the proportion of diamonds are between .3 and 1.08 is 0.74683.

## 7. Make boxplots of the diamond price for each cut and discuss any conclusions you can draw from the appearance of the boxplot
###create a summary statistic that display the price in different cut levl
###exchange the positions of price and cut level in summary statistic
###create the box plot to display the price in different level

```python
#use groupby function to identify the cut level and get the summary statistic of price in different cut level
group  = df.groupby('cut')[['price']].describe()
print(group)
```

```python
#exchange the position of cut level and price
group.T
```

```python
#use the boxplot that reflects all cut level 
group.T.boxplot()
```

According to the box plot, the most expensive diamonds is Ideal cut level. In addition, lowest price for different level cut are closed. Premium cut level has the largest range of price.
