## 1. What proportion of diamonds are between .30 and 1.08 carats?
#define the pandas and numpy
#read the url from github
#create a datafarm that calculates the proportion of diamonds in column 'carat' that satisfy the requirement

```python
#definde pandas and numpy
import pandas as pd
import numpy as np
# use read_csv function to load the csv from spcific websites.
df = pd.read_csv('https://raw.githubusercontent.com/mandy941/week6/master/diamonds.csv')
#calculate the value of column 'carat' that between 0.3 and 1.08 divides the total numble of rows.
porp = pd.value_counts((df['carat'] <= 1.08) & (df['carat'] >= 0.30)) / df.shape[0]
print(porp)
```

According to the result, the proportion of diamonds are between .3 and 1.08 is 0.74683.
